```python
from ase.build import bulk
from atomistics.calculators import get_potential_by_name, calc_molecular_dynamics_nvt_with_lammpslib
from pylammpsmpi import LammpsASELibrary
from executorlib import FluxJobExecutor
```

```python
structure = bulk("Ti")
```

```python
potential_dataframe = get_potential_by_name(
    potential_name="2016--Mendelev-M-I--Ti-3--LAMMPS--ipr1"
)
```

```python
with FluxJobExecutor(resource_dict={"cores": 2}) as exe:
    lmp = LammpsASELibrary(executor=exe, cores=2)
    result_dict = calc_molecular_dynamics_nvt_with_lammpslib(
        structure=structure,
        potential_dataframe=potential_dataframe,
        lmp=lmp
    )
    lmp.close()
```
